2.- Realice la extracción del guion de una película (en *.csv o *.txt). (3 puntos). Creo que solamente es el título del guión.

```python
# Librerías
import pandas as pd
import os

# para eliminar warning en python
import warnings
warnings.simplefilter('ignore')
```

```python
# Leer el archivo CSV en un DataFrame
datos = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/peliculas_scraped.csv")
```

```python
datos.head(2)
```

```python
# Seleccionar las columnas 'movie_title' y 'title' y crear un nuevo DataFrame
datos_seleccionados = datos[['movie_title', 'title']]
datos_seleccionados.head(2)
```

```python
# Renombrar 'movie_title' a 'Titulo de la pelicula' y 'title' a 'Guion'
datos_seleccionados.rename(columns={'movie_title': 'Titulo de la pelicula', 'title': 'Guion'}, inplace=True)
datos_seleccionados.head(2)
```

```python
datos_seleccionados["Guion"].unique()
```

```python
# Guardando el DataFrame como guion.csv
datos_seleccionados.to_csv("guion.csv", index=False)
```
