```python
# Librerías a utilizar
import requests
from bs4 import BeautifulSoup
import pandas as pd

# USER AGENT PARA PROTEGERNOS DEL BANEOS DEL SERVIDOR
headers= {
    "user-agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}

url= 'https://en.wikipedia.org/wiki/%C3%8Dndice_de_Precio_Selectivo_de_Acciones'

# Realizar la solicitud HTTP
respuesta = requests.get(url, headers=headers)
```

```python
# Esto permite analizar y extraer información del texto de una manera estructurada.
soup = BeautifulSoup(respuesta.text)
```
