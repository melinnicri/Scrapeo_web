```python
# Librer√≠as
import pandas as pd
import os

# para eliminar warning en python
import warnings
warnings.simplefilter('ignore')
```

```python
# Leer el archivo CSV en un DataFrame
datos = pd.read_csv("/content/peliculas_scraped.csv")
```

```python
datos.head(2)
```

```python
# Pedido dec√≠a "extraiga las pel√≠culas y sus correspondientes links"
# Seleccionar las columnas 'movie_title' y 'href' y crear un nuevo DataFrame
datos_seleccionados = datos[['movie_title', 'href']]
datos_seleccionados.head(2)
```

```python
# Renombrar 'movie_title' a 'T√≠tulo de la pel√≠cula' y 'href' a 'Enlace_web'
datos_seleccionados.rename(columns={'movie_title': 'T√≠tulo de la pel√≠cula', 'href': 'Enlace_web'}, inplace=True)
datos_seleccionados.head(2)
```

```python
# Guardando el DataFrame como pedido.csv
datos_seleccionados.to_csv("pedido.csv", index=False)
```
